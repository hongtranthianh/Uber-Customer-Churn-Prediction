`5.2`. Find the number of trips that were given specific rating score in the scale 0-5 for driver and rider

```python
driver_score_df = df['driver_score'].value_counts().reset_index()
driver_score_df.columns=['score','driver']

rider_score_df = df['rider_score'].value_counts().reset_index()
rider_score_df.columns=['score','rider']

# Merge two df to be ready for a bar chart
score_df = driver_score_df.merge(rider_score_df, how='inner', on='score').sort_values('score')

# group bar chart
score_df.set_index('score').plot(kind='bar', figsize=(6,3), rot=0)
plt.title('The number of trips given specific rating score for rider and driver')
plt.xlabel('rating score')
plt.ylabel('trip count');
```

`5.3`. When was the peak period in a month/weekday/hour?

```python
df['date'] = df['start_at'].dt.normalize()
df['month'] = df['start_at'].apply(lambda x: x.month)
df['month_year'] = df['start_at'].dt.to_period('M')
df['day_of_month'] = df['start_at'].apply(lambda x: x.day)
df['day_of_week'] = df['start_at'].apply(lambda x: x.strftime('%A'))
df['hour'] = df['start_at'].apply(lambda x: x.hour)
```

```python
day_of_month_df = df.groupby(['date','day_of_month'])['journey_id'].count().groupby('day_of_month').mean().reset_index()
day_of_month_df.plot(kind='line', x='day_of_month', y='journey_id', legend=False, figsize=(9,4))
plt.title('Average number of trips of each day in a month')
plt.xlabel('day of month')
plt.ylabel('Trip count')
plt.xticks(day_of_month_df['day_of_month']); #to show all value in x_axis
```

```python
weekday_order = category_order_list=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
day_of_week_df = df.groupby(['date','day_of_week'])['journey_id'].count().groupby('day_of_week').mean().reset_index()
day_of_week_df["day_of_week"] = pd.Categorical(day_of_week_df["day_of_week"], categories = weekday_order)
day_of_week_df.sort_values(by = "day_of_week", inplace=True)
day_of_week_df.plot(kind='line', x='day_of_week', y='journey_id', legend=False, figsize=(7,4))
plt.title('Average number of trips of each day in a week')
plt.xlabel('day of week')
plt.ylabel('Trip count');
```

```python
hour_df = df.groupby(['date','hour'])['journey_id'].count().groupby('hour').mean().reset_index()
hour_df.plot(kind='line', x='hour', y='journey_id', legend=False, figsize=(8,4))
plt.title('Average number of trips of each hour in a day')
plt.xlabel('hour')
plt.ylabel('Trip count')
plt.xticks(hour_df['hour']); #to show all value in x_axis
```

```python
end_state_hour_df = df.groupby(['date','hour','end_state'])['journey_id'].count()\
                      .groupby(['hour','end_state']).mean().unstack()\
                      .fillna(0)\
                      .drop(['no show','not found'], axis=1)
# Line chart
end_state_hour_df.plot(kind='line', rot=0, figsize=(8,4))
plt.title('Average number of trips of each hour in a day as per end state')
plt.xlabel('hour')
plt.ylabel('Trip count')
plt.xticks(end_state_hour_df.index); #to show all value in x_axis
```

It seems the peak hour is 1 pm and 11pm, it's a little bit weird compared to normal peak hour between 7am and 10am as we commonly know. The time zone used here maybe UTC and it didn't reflect time zone in Peru (Lima). So we will change the UTC into Peru timezone and let's see the resut again.

```python
# Convert timezone
df['start_at_ptz'] = df['start_at'].dt.tz_localize('Etc/GMT+1').dt.tz_convert('America/Lima')
df['date_ptz'] = df['start_at_ptz'].dt.normalize()
df['month_ptz'] = df['start_at_ptz'].apply(lambda x: x.month)
df['month_year_ptz'] = df['start_at_ptz'].dt.to_period('M')
df['day_of_month_ptz'] = df['start_at_ptz'].apply(lambda x: x.day)
df['day_of_week_ptz'] = df['start_at_ptz'].apply(lambda x: x.strftime('%A'))
df['hour_ptz'] = df['start_at_ptz'].apply(lambda x: x.hour)
```

```python
hour_df1 = df.groupby(['date','hour_ptz'])['journey_id'].count().groupby('hour_ptz').mean().reset_index()
hour_df1.plot(kind='line', x='hour_ptz', y='journey_id', legend=False, figsize=(8,4))
plt.title('Average number of trips of each hour in a day')
plt.xlabel('hour (Peru timezone)')
plt.ylabel('Trip count')
plt.xticks(hour_df1['hour_ptz']); #to show all value in x_axis
```

```python
hour_df2 = df.groupby(['date','hour_ptz','day_of_week_ptz'])['journey_id'].count()\
                      .groupby(['hour_ptz','day_of_week_ptz']).mean().unstack()\
                      .fillna(0)
hour_df2 = hour_df2[['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']]
# Line chart
hour_df2.plot(kind='line', rot=0, figsize=(8,4))
plt.title('Average number of trips of each hour in a day as per weekday')
plt.xlabel('hour (Peru timezone)')
plt.ylabel('Trip count')
plt.xticks(hour_df2.index); #to show all value in x_axis
```

```python
hour_df3 = df.groupby(['date','hour_ptz','end_state'])['journey_id'].count()\
                      .groupby(['hour_ptz','end_state']).mean().unstack()\
                      .fillna(0)\
                      .drop(['no show','not found'], axis=1)
# Line chart
hour_df3.plot(kind='line', rot=0, figsize=(8,4))
plt.title('Average number of trips of each hour in a day as per end state')
plt.xlabel('hour (Peru timezone)')
plt.ylabel('Trip count')
plt.xticks(end_state_hour_df.index); #to show all value in x_axis
```

Now the peak hour is between 8-10am, and after 5pm.

```python
day_of_month_df = df.groupby(['date','day_of_month_ptz'])['journey_id'].count().groupby('day_of_month_ptz').mean().reset_index()
day_of_month_df.plot(kind='line', x='day_of_month_ptz', y='journey_id', legend=False, figsize=(9,4))
plt.title('Average number of trips of each day in a month')
plt.xlabel('day of month (Peru timezone)')
plt.ylabel('Trip count')
plt.xticks(day_of_month_df['day_of_month_ptz']); #to show all value in x_axis
```

```python
weekday_order = category_order_list=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
day_of_week_df = df.groupby(['date','day_of_week_ptz'])['journey_id'].count().groupby('day_of_week_ptz').mean().reset_index()
day_of_week_df["day_of_week_ptz"] = pd.Categorical(day_of_week_df["day_of_week_ptz"], categories = weekday_order)
day_of_week_df.sort_values(by = "day_of_week_ptz", inplace=True)
day_of_week_df.plot(kind='line', x='day_of_week_ptz', y='journey_id', legend=False, figsize=(7,4))
plt.title('Average number of trips of each day in a week')
plt.xlabel('day of week (Peru timezone)')
plt.ylabel('Trip count');
```

`5.4`. How did the Uber service evolve over the year 2010?

`5.4.1`. Number of trips

```python
trip_end_state_df = df.groupby(['month_year_ptz','end_state'])['journey_id'].count().unstack()
trip_end_state_df.fillna(0, inplace=True)
trip_end_state_df.drop(['no show','not found'], axis=1, inplace=True)
trip_end_state_df.plot(kind='line', rot=0, figsize=(6,4))
plt.title('Number of trips over the year 2010')
plt.xlabel('Month year')
plt.ylabel('Trip count');
```

`5.4.2`. Number of new customers

```python
new_riders_df1 = df.groupby('user_id')['month_year_ptz'].min().reset_index()\
                  .groupby('month_year_ptz')['user_id'].count().reset_index()
new_riders_df1.columns=['month_year','All end states']
new_riders_df2 = df[df['end_state']=='drop off'].groupby('user_id')['month_year'].min().reset_index()\
                  .groupby('month_year')['user_id'].count().reset_index()
new_riders_df2.columns=['month_year','Drop off trips only']
# Merge two df to be ready for a group line chart
new_riders_df = new_riders_df1.merge(new_riders_df2, how='inner', on='month_year')

# group line chart
new_riders_df.set_index('month_year').plot(kind='line', rot=0, figsize=(6,4))
plt.title('Number of new customers over the year 2010')
plt.ylabel('New customers');
```

`5.5`. Cancellation fee

The ride can be cancelled by either the driver or the rider after booking. According to Uber policy [here](https://www.uber.com/en-KE/blog/cancellation-fees/), riders can be charged cancellation fee depending on how far the driver has already driven, how long it took the driver to arrive at the pickup location, and the time your driver has been waiting, or canceling trips more than once in a short period of time.

We will count the trips that were canceled by riders with positive `price` as being charged cancellation fee.

```python
# How many trips were canceled by riders?
df[df['end_state']=='rider cancel']['journey_id'].count()
```

```python
# How many trips were canceled by riders and being charge cancellation fee?
df.loc[(df['end_state']=='rider cancel')&(df['price']>0)]['journey_id'].count()
```
